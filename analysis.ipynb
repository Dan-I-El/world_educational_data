```python
for column in df:
    if df[column].dtype == np.int64 and df[column].max() < 128:
        df[column] = df[column].astype(np.int8, copy=None)
        print(df[column].dtype)
    elif df[column].dtype == np.float64 and df[column].max() < 200:
        df[column] = df[column].astype(np.float16, copy=None)
```

```python
final_df = pd.merge(df_countries, df_c, left_on="countries_and_areas", right_on="country")
final_df.sample()
```

```python
px.choropleth(
    data_frame=final_df,
    locations="iso_alpha",
    hover_name="countries_and_areas",
    color="oosr_pre0primary_age_male",
    width=1400,
    height=600,
    center={"lat": 45.6, "lon": -73.5},
    title="Unemployment rate per country",
    color_discrete_sequence=px.colors.qualitative.Set1
)
```
