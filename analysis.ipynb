```python
for column in df:
    if df[column].dtype == np.int64 and df[column].max() < 128:
        df[column] = df[column].astype(np.int8, copy=None)
        print(df[column].dtype)
    elif df[column].dtype == np.float64 and df[column].max() < 200:
        df[column] = df[column].astype(np.float16, copy=None)
```
